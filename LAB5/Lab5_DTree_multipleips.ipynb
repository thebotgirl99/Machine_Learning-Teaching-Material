```python
#import all the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/content/diabetes2.csv') #read the dataset from the file location in google colab
data = dataset[['Glucose', 'BMI', 'Insulin', 'Outcome']] #choose your feature and label
x, y =np.array(data[['Glucose', 'BMI', 'Insulin']]).reshape(-1,3), np.array(data['Outcome']).reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25) #split the data into training and testing data
```

```python
clf = tree.DecisionTreeClassifier(max_depth=4) #specify the depth (height/no.of branches) of your tree
clf = clf.fit(x_train, y_train) #train your data
```

```python
clf.predict(x_test)  #predict the new (test) data using your trained model
```

```python
clf.score(x_test, y_test) #predict the accuracy of your model
```

```python
feat = ['Glucose', 'BMI', 'Insulin'] #specify the names of your feature and class (0 or 1)
labl = ['no', 'yes']
fig = plt.figure(figsize=(25,20)) #plot your decision tree. Gini: probability of a variable being wrongly classified when it is randomly chosen
tree.plot_tree(clf, feature_names= feat,  
                   class_names= labl,
                   filled=True)
```
