Whenever your dataset has strings instead of numericals, we call it as **'categorical'** data
A machine learning model can only accept **numeric** data, which is why we have to convert our categorical data to numerical data

```python
import numpy as np  #import the necessary functions
import pandas as pd

dataset = pd.read_csv('/content/measurements.csv') #read your dataset 
dataset.head  #display first 5 and last 5 values of your dataset
```

```python
gas = pd.get_dummies(dataset['gas_type'])   #convert the categorical values to numerical values '0' or '1'
data_new = pd.concat([dataset, gas], axis=1)  #join the above column with your origianl dataset
data_new  #display your new dataset
```

```python
data_new.corr()  #if you do not want to visualize your correlation matrix use this method
```

```python
data_new.describe()  #use this function to learn the statistics of your dataset
```
